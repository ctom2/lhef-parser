Import of the methods and classes from `lhefparser.py` file.

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
from lhefparser import *
```

Reading the data from LHEF file.

```python
p = ReadLHEF('./LHEF_files/evrecout.dat')
```

```python
print(len(p.events))
```

Printing the number of particles in one event.

```python
print(p.events[0].nparticles)
```

---

```python
energy_a = p.init.energy_a
energy_b = p.init.energy_b
```

```python
def getXi(energy_beam, energy_p):
    return (energy_beam - energy_p)/energy_beam
```

```python
Ep1 = GetData(p, [0], 'E', 'eV') # energy of proton 1
Ep2 = GetData(p, [1], 'E', 'eV') # energy of proton 2
```

```python
xi_p1 = []
xi_p2 = []

for data in Ep1:
    xi_p1.append(getXi(energy_a, data))
    
for data in Ep2:
    xi_p2.append(getXi(energy_b, data))
```

```python
plt.hist(xi_p1, bins=25, histtype='step', align='right', color='red', linewidth=1.8, label='photon 1')

plt.axvline(0.02, color='black', linewidth=1)
plt.axvline(0.10, color='black', linewidth=1)

label = 'Proton 1 relative energy loss, ξ'
plt.xlabel(label)
plt.ylabel('Events')

plt.show()
```

```python
plt.hist(xi_p1, bins=50, histtype='step', align='right', color='red', linewidth=1.8, label='photon 1')

plt.axvline(0.02, color='black', linewidth=1)
plt.axvline(0.10, color='black', linewidth=1)

label = 'Proton 1 relative energy loss, ξ'
plt.xlabel(label)
plt.ylabel('Events')

plt.axis([0.01, 0.11, 0, 1500])

plt.show()
```

```python
plt.hist(xi_p1, bins=25, histtype='step', align='right', color='blue', linewidth=1.8, label='photon 2')
plt.axvline(0.02, color='black', linewidth=1)
plt.axvline(0.10, color='black', linewidth=1)

label = 'Proton 2 relative energy loss, ξ'
plt.xlabel(label)
plt.ylabel('Events')

plt.show()
```

```python
plt.hist(xi_p2, bins=50, histtype='step', align='right', color='blue', linewidth=1.8, label='photon 2')

plt.axvline(0.02, color='black', linewidth=1)
plt.axvline(0.10, color='black', linewidth=1)

label = 'Proton 2 relative energy loss, ξ'
plt.xlabel(label)
plt.ylabel('Events')

plt.axis([0.01, 0.11, 0, 1500])

plt.show()
```

```python
n = 0

for event in p.events:
    energy = event.particles[0].E
    xi = getXi(energy_a, energy)
    if xi > 0.02 and xi < 0.10:
        n = n + 1

print(n)
```

```python
n = 0

for event in p.events:
    energy = event.particles[1].E
    xi = getXi(energy_b, energy)
    if xi > 0.02 and xi < 0.10:
        n = n + 1

print(n)
```
